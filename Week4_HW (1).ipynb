4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python

# Generate random heatmap data (positive correlation)
np.random.seed(0)
n = 200
data1 = np.random.rand(n)
data2 = 2 * data1 + np.random.rand(n)  # Positive correlation

# Create a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(data1, data2, c='cyan', marker='o', label='Data Points')

#plotting the graph
plt.xlabel('Data Point 1')
plt.ylabel('Data Point 2')
plt.title('Scatter Plot QUESTION 4')
plt.legend()
plt.grid(True)
plt.show()
```

5. Now, create a scatter plot of two heatmap entries that appear to have negative correlation.

```python

np.random.seed(0)
m = 200
data1 = np.random.rand(m)
data2 = -2 * data1 + np.random.rand(m)  # Negative correlation

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data1, data2, c='red', marker='o', label='Data Points')

# Add labels and a title
plt.xlabel('Data 1')
plt.ylabel('Data 2')
plt.title('Scatter Plot of Two Negatively Correlated Heatmap Entries')

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```
